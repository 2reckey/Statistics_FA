#ЗАДАЧА 4.1

```python
import numpy as np
sin_values = np.sin(np.arange(1, 101))
sin_mean = np.mean(sin_values)
print('Эмпирическое среднее совокупности sin:', sin_mean)
sin_std = np.std(sin_values)
print('Эмпирическое стандартное отклонение совокупности sin:', sin_std)
```

```python
sin_moment4 = np.mean((sin_values - sin_mean) ** 4)
sin_excess = sin_moment4 / (sin_std ** 4) - 3
print('Четвертый эмпирический центральный момент совокупности sin:', sin_moment4)
print('Эмпирический эксцесс совокупности sin:', sin_excess)
```

```python
import matplotlib.pyplot as plt
x= np.arange(1,101)
y = np.unique(np.sin(x))
plt.plot(x, y)
plt.title('График совокупности sin')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
cos_values = np.unique(np.cos(np.arange(1, 101)))
plt.plot(x, y, label='sin')
plt.plot(x, cos_values, label='cos')
plt.title('Графики совокупностей: sin и cos')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
sin_cos_corr = np.corrcoef(sin_values, cos_values)[0,1]
print('Эмпирический коэффициент корреляции между совокупностями sin и cos:', sin_cos_corr)
```

#ЗАДАЧА 4.2

```python

prime = [3, 5, 7, 11]
i = 13;
n = 99992;
while i < n:
    if pow(2, i - 1, i) == 1:
        for pr in prime:
            if i % pr == 0: break
            if pr > i // pr:
                prime.append(i)
                break
    i += (2, 4)[i % 6 == 1]

prime_mean = np.mean(prime)
prime_moment3 = np.mean((prime - prime_mean) ** 3)
prime_std = np.std(prime)
print("Ассиметрия простых чисел:", prime_moment3 / prime_std ** 3)
```

```python
for i in range(0, len(prime) * 2, 2): prime.insert(0, -prime[i])
prime_mean = np.mean(prime)
prime_std = np.std(prime)
prime_moment4 = np.mean((prime - prime_mean) ** 4)
print("Эксцесс простых чисел:", prime_moment4 / prime_std ** 4 - 3)
```

#ЗАДАЧА 4.3

```python
def avg(list, a):
    if a == 0:
        return np.exp(np.mean(np.log(list)))
    return np.mean(np.power(list, float(a))) ** (1 / a)


prime = list(filter(lambda x: x > 10 and x // 100 == 0, prime))

print("Среднее геометрическое простых двузначных чисел:", avg(prime, 0))
print("Среднее гармоническое простых двузначных чисел:", avg(prime, -1))
print("Среднее кубическое простых двузначных чисел:", avg(prime, 3))
```
